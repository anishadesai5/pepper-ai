```python
# Setup vars and get list of running and installed behaviors
# NOTE : this might error if the package is already installed.
from naoqi import ALProxy

import os
from dotenv import load_dotenv

# Load environment variables from the .env file
load_dotenv()

# Access the environment variables
PIP = os.getenv("PIP")
PPORT = int(os.getenv("PPORT"))

behavior_manager = ALProxy("ALBehaviorManager", PIP, PPORT)

# Start behavior
behavior_name_to_start = "starter-talbetpage1"


# Check if the behavior is running before attempting to stop it
if behavior_name_to_start in behavior_manager.getInstalledBehaviors():
    # Stop all the behaviors
    behavior_manager.stopAllBehaviors()
    #start a new one
    behavior_manager.startBehavior(behavior_name_to_start)
    print("Behavior '{}' Started.".format(behavior_name_to_start))
else:
    print("Behavior '{}' is not available.".format(behavior_name_to_start))
```

```python
# Uninstall the Orignal default package (which occasionally gets reinstalled by the system)
from naoqi import ALProxy

# Replace with the correct IP and port
robot_ip = os.getenv("PIP")
robot_port = 9559

# Connect to the package manager on the robot
package_manager = ALProxy("PackageManager", robot_ip, robot_port)

package_name = "ht_cms_1_5"  # replace with the actual package name
package_manager.removePkg(package_name)


installed_packages = package_manager.packages()
print(installed_packages)
if package_name not in installed_packages:
    print("Package %s has been successfully uninstalled.", package_name)
else:
    print("Failed to uninstall package %s. , ", package_name)
```

```python
#SLEEP PEPPER

from naoqi import ALProxy

# Replace these with the IP and port of your Pepper robot


def put_pepper_to_sleep():
    try:
        # Create a proxy to ALMotion
        motion_proxy = ALProxy("ALMotion", PIP, PPORT)
        
        # Set Pepper to rest
        motion_proxy.rest()
        print("Pepper is now asleep.")
        
    except Exception as e:
        print("Error when putting Pepper to sleep: ", e)

put_pepper_to_sleep()
```

```python
# WAKE PEPPER
from naoqi import ALProxy


def wake_pepper_up():
    try:
        # Create a proxy to ALMotion
        motion_proxy = ALProxy("ALMotion", PIP, PPORT)
        
        # Wake Pepper up
        motion_proxy.wakeUp()
        print("Pepper is now awake.")
        
    except Exception as e:
        print("Error when waking Pepper up: ", e)

wake_pepper_up()
```
